```python
start = datetime.date(2006, 10, 31)
end = datetime.date(2019, 10, 31)
start
```

```python
def add_month(dat, x):
    while x != 0:
        if (dat + datetime.timedelta(1)).month != dat.month:
            dat += datetime.timedelta(1)
            x -= 1
            if x == 0:
                while (dat + datetime.timedelta(1)).month == dat.month:
                    dat += datetime.timedelta(1)
        else:
            dat += datetime.timedelta(1)
    return dat
add_month(start, 12)
```

```python
lst = []
while start != end:
    lst.append(start)
    start = add_month(start, 3)
lst
```

```python
new_df = pd.DataFrame(columns = ('Date', 'Open', 'High', 'Low', 'Adj Close', 'Volume', 'EPS'))
for i in range(len(eps) - 1):
    temp = df2[(df2['Date'] >= eps['index'][i]) & (df2['Date'] < eps['index'][i+1])]
    temp['EPS'] = eps['EPS'][i]
    new_df = new_df.append(temp)
new_df
```

```python
temp = df2[(df2['Date'] >= eps['index'][51])]
temp['EPS'] = eps['EPS'][51]
new_df = new_df.append(temp)
new_df
```

```python
trace0 = go.Scatter(
    x = new_df['Date'], y = new_df['EPS'], name = 'EPS', yaxis = 'y2'
)
trace1 = go.Scatter(
    x = new_df['Date'], y = new_df['Volume'], name = 'Volume'
)
trace2 = go.Scatter(
    x = new_df['Date'], y = new_df['Adj Close'], name = 'CSCO Price'
)

data = [trace0, trace1, trace2]
updatemenus = list([
    dict(type = "buttons",
        active = -1,
         buttons = list([
             dict(label = 'Price',
                 method = 'update',
                 args = [{'visible': [True, False, True]},
                        {'title': 'CSCO Share Price'}]),
             dict(label = 'Volume',
                 method = 'update',
                 args = [{'visible': [True, True, False]},
                        {'title': 'CSCO Trading Volume'}]),
         ])
        )
])

    
layout = go.Layout(
    title = 'CSCO', yaxis = dict(title = 'Price or Volume'), yaxis2=dict(title = 'EPS', overlaying='y', \
                                                                           side='right'), updatemenus  = updatemenus)
fig = go.Figure(data=data, layout=layout)
plotly.offline.iplot(fig)
```
